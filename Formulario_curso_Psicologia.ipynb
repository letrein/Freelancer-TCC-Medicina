## Sobre os dados:

Os dados são oriundos de um questionário de uma unioversidade sobre questões dos alunos de vários cursos. 

## Colunas

As colunas possuem como nome uma palavra chave para as seguintes questões:


O Plano de Desenvolvimento Institucional – PDI é o documento que identifica a FAPAR/FAC, no que diz respeito à sua filosofia de trabalho, à missão a que se propõe, às diretrizes pedagógicas

A forma que o Projeto Pedagógico do Curso (PPC) está estruturada consigo acompanhar as atividades de ensino, com carga horária adequadas as atividades previstas, contribuindo, assim, para minha formação

As disciplinas em EAD (ensino a distância) estão bem estruturadas no ambiente virtual (AVA) no portal do aluno, com aulas virtuais gravadas, materiais de apoio, materiais complementares e interatitivas

As atividades supervisionadas (APS) são atividades de extensão realizadas e desenvolvidas sob orientação e avaliação de um professor. Estão previstas no PPC do Curso com carga horária definida

A metodologia de ensino e extensão do curso atende ao desenvolvimento de conteúdos, às estratégias de aprendizagem para assim, contribuir a autonomia do discente, em uma relação teoria-prática

A gestão do curso, é realizada considerando a auto avaliação institucional e o resultado das avaliações externas (ENADE, avalições in loco do MEC) como insumo para aprimoramento contínuo do planej


As atividades de tutoria atendem às demandas didático-pedagógicas da estrutura curricular, compreendendo a mediação pedagógica junto aos discentes, inclusive em momentos presenciais 

O coordenador atende à demanda existente, considerando a gestão do curso, a relação com os docentes e discentes, com tutores e equipe multidisciplinar favorecendo a integração e a melhoria continua

Evidencio políticas de atendimento discente em ações de acolhimento par minha permanência na FAPAR/FAC: com acessibilidade metodológica e instrumental, monitoria, nivelamento 

A comunicação da FAPAR/FAC é efetiva e comprometida com a missão da IES. A informação divulgada inclui os aspectos que dizem respeito às atividades da instituição (cursos ofertados e modalidades

```python
import pandas as pd

curso = "Psicologia"
caminho = "H:/Meu Drive/Nilce/Psicologia/form_psicologia.csv"
# Leitura do arquivo CSV com tratamento de erros
df = pd.read_csv(caminho, 
                 sep=',', encoding='utf-8')  
# Exibir as primeiras linhas do DataFrame
df.head()
```

## Arrumando variáveis de string para numérico


```python
replace_map = {
    'concordo plenamente\xa0': 5,
    'concordo parcialmente': 4,
    'nem concordo, nem discordo': 3,
    'discordo parcialmente': 2,
    'discordo plenamente\xa0': 1,
    'não sei opinar': 0
}

df.replace(replace_map, inplace=True)
```

```python
df.head
```

```python
# Dicionário com os novos nomes das colunas
novos_nomes_colunas = {
    'PDI ': 'Plano de Desenvolvimento Institucional',
    'PCC': 'Projeto Pedagógico do Curso',
    'EAD': 'Ensino a Distância',
    'APS': 'Atividades Supervisionadas',
    'Extensao': 'Extensão',
    'ENADE': 'Avaliações Externas',
    'Tutoria': 'Atividades de Tutoria',
    'Coordenador': 'Coordenação do Curso',
    'Acolhimento': 'Políticas de Acolhimento',
    'Comunicação': 'Comunicação Institucional',
    'Nivel_satisfacao': 'Nível de Satisfação Geral'
}

# Renomear as colunas do DataFrame
df.rename(columns=novos_nomes_colunas, inplace=True)

# Exibir as primeiras linhas do DataFrame para verificar as mudanças
df.head()
```

```python
df_novo = pd.read_csv(caminho_novo)
df_novo = df_novo.drop(columns=['Comentario'])
df_novo.head()
```

## Plotagem de gráficos


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for coluna in df_novo.columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df_novo, x=coluna, palette="muted")
    plt.title(f"Distribuição de Respostas: {coluna} - Curso de {curso}")
    plt.xlabel("Respostas")
    plt.ylabel("Frequência")
    plt.ylim(0,20)
    total = df_novo[coluna].value_counts().sum()
    for p in plt.gca().patches:
        altura = p.get_height()
        plt.gca().text(p.get_x() + p.get_width() / 2, altura + 0.5, f"{altura / total:.1%}", ha='center')
    
    plt.show()
```

```python
media_respostas = df_novo.mean()
plt.figure(figsize=(15, 6))
sns.lineplot(data=media_respostas, marker='o', color="blue")
plt.title("Média das Respostas por Questão")
plt.xlabel("Questões")
plt.ylabel("Média")
plt.xticks(rotation=20)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=df_novo, orient="h", palette="pastel")
plt.title("Distribuição Comparativa das Respostas por Questão")
plt.xlabel("Pontuação")
plt.show()
```
