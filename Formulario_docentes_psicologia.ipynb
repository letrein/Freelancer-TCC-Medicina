## Sobre os dados:

Os dados são oriundos de um questionário de uma unioversidade sobre os professores do curso de graduacao


## Colunas

As colunas possuem como nome uma palavra chave para as seguintes questões:


a) O professor apresentou e cumpriu o plano de ensino apresentado em sala de aula.
b) O professor demonstrou domínio dos conteúdos.
c) O professor trouxe o conteúdo para exemplos da realidade do aluno com o objetivo de simplificar o aprendizado
d) O professor utilizou recursos (quadros, Datashow, cartazes, textos, instrumentos de acordo com a disciplina: modelos, tubo de ensaio, microscópio) e procedimentos didáticos adequados.
e) O professor trabalhou com clareza e objetividade.
f) O professor disponibilizou tempo para atender os discentes em sala de aula ou fora da sala de aula, pessoalmente e/ou à distância.
g) O professor foi assíduo e pontual.
h) O professor realizou avaliações compatíveis com o que foi trabalhado nas aulas.
i) O professor analisou com os discentes os resultados das avaliações, fez a devolutiva.
j) O professor manteve atitudes de respeito e cortesia.

# ANÁLISE DOS DADOS

```python
import pandas as pd

curso = "Psicologia"
caminho = "H:/Meu Drive/Nilce/Psicologia/form_docente_psico.csv"
# Leitura do arquivo CSV com tratamento de erros
df = pd.read_csv(caminho, 
                 sep=',', encoding='utf-8')  
# Exibir as primeiras linhas do DataFrame
df.head()
```

## Arrumando variáveis de string para numérico


```python
df['Plano de ensino'].unique()
```

```python
replace_map = {
    'concordo plenamente Score: 4,1 - 5\xa0': 5,
    'concordo parcialmente\xa0 Score: 3,1 - 4': 4,
    'nem concordo e nem discordo Score: 2,1 -\xa0 3\xa0': 3,
   'discordo parcialmente\xa0 Score: 1,1- 2': 2,
   'discordo totalmente , Score: 0-1': 1,
    'não sei opinar': 0
}

df.replace(replace_map, inplace=True)
```

```python
df.head
```

```python
# Salvar o novo dataframe com o nome semelhante ao original, acrescido de '_numerico'
caminho_novo = "H:/Meu Drive/Nilce/Psicologia/form_docentes_psico_numerico.csv"
df.to_csv(caminho_novo, index=False)
```

```python
df_professores = pd.read_csv(caminho_novo)
df_professores.head()
```

## Plotagem de gráficos


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
   
for professor in df_professores['DOCENTE'].unique():
    df_professor = df_professores[df_professores['DOCENTE'] == professor]
    df_professor_mean = df_professor.mean(numeric_only=True)
    
    plt.figure(figsize=(7, 5))
    sns.barplot(y=df_professor_mean.index, x=df_professor_mean.values, palette="pastel")
    plt.title(f"Média das Avaliações - {professor}")
    plt.ylabel("Critérios de Avaliação")
    plt.xlabel("Média")
    plt.xlim(0, 5.5)
    #adicionando valor total de avaliações no gráfico
    for i, v in enumerate(df_professor_mean.values):
        plt.text(v, i, str(round(v, 2)), color='black', ha="left")
    plt.show()
```

```python
# Agrupar os dados por professor e calcular as médias das avaliações
df_professores_grouped = df_professores.groupby('DOCENTE').mean(numeric_only=True).reset_index()

# Exibir as primeiras linhas
df_professores_grouped.head()

```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_professores_grouped.set_index('DOCENTE').T, annot=True, cmap="viridis", fmt=".1f")
plt.title("Mapa de Calor das Avaliações por Professor")
plt.xlabel("Professor")
plt.ylabel("Critérios de Avaliação")
plt.show()
```

```python
df_means = df_professores.groupby('DOCENTE').mean(numeric_only=True)
df_means.T.plot(kind='bar', figsize=(15, 6), colormap='Set2')
plt.title("Comparação de Avaliações por Professor")
plt.ylabel("Pontuação Média")
plt.xlabel("Critérios de Avaliação")
plt.xticks(rotation=20)
plt.legend(title="Professores", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
